```python
secret_key = config['SECRET_KEY']
```

```python
fred = Fred(api_key=secret_key)
```

# 4. pull and join Multiple Data Series

```python
unemp_df = fred.search('unemployment state', filter=('frequency', 'Monthly'))
unemp_df = unemp_df.query('seasonal_adjustment == "Seasonally Adjusted" and units == "Percent"')
unemp_df = unemp_df.loc[unemp_df['title'].str.contains('Unemployment Rate')]
```

```python
all_results = []
for i in unemp_df.index:
    result = fred.get_series(i)
    result = result.to_frame(name=i)
    all_results.append(result)
unem_df = pd.concat(all_results, axis=1).drop(['M08311USM156SNBR'], axis=1)
```

```python
unemp_states = unem_df.drop('UNRATE', axis=1)
```

```python
unemp_states = unemp_states.dropna()
```

```python
fig = px.line(unemp_states)
```
